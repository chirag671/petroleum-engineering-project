```python
Q=float(input('enter constant flow rate'))
B=float(input('enter formation volume factor'))
muo=float(input('enter viscosity'))
k=float(input('enter permeability'))
Pi=float(input('enter initial reservior pressure'))
Ct=float(input('enter formation compresibility'))
rw=float(input('enter well bore radius'))
h=float(input('enter thickness'))
```

```python
phi=float(input('enter the porosity'))
t1=float(input('enter time'))
t2=float(input('enter time'))
t3=float(input('enter time'))
```

```python
import pandas as pd
pressure_profil=pd.DataFrame({'r(inner radius)':[0.25,5,10,50,100,500,1000,1500,2000,2500]})
pressure_profil
```

```python
pressure_profil['x1']=(948*muo*phi*Ct)*(pressure_profil['r(inner radius)']**2)/(k*t1)
pressure_profil['x2']=(948*muo*phi*Ct)*(pressure_profil['r(inner radius)']**2)/(k*t2)
pressure_profil['x3']=(948*muo*phi*Ct)*(pressure_profil['r(inner radius)']**2)/(k*t3)
pressure_profil
```

```python
pressure_profil['E(-x1)']=np.log(1.78*pressure_profil['x1'])
pressure_profil['E(-x2)']=np.log(1.78*pressure_profil['x2'])
pressure_profil['E(-x3)']=np.log(1.78*pressure_profil['x3'])
pressure_profil
```

```python
pressure_profil.loc[5:,'E(-x1)']=0
pressure_profil.loc[8:,'E(-x2)']=0
pressure_profil.loc[9,'E(-x3)']=0
pressure_profil
```

```python
pressure_profil['Pr1']=(70.6*Q*muo*B)/(k*h)*pressure_profil['E(-x1)']+Pi
pressure_profil['Pr2']=(70.6*Q*muo*B)/(k*h)*pressure_profil['E(-x2)']+Pi
pressure_profil['Pr3']=(70.6*Q*muo*B)/(k*h)*pressure_profil['E(-x3)']+Pi
pressure_profil
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(pressure_profil['r(inner radius)'],pressure_profil['Pr1'],label='1 hour',color='g')
plt.plot(pressure_profil['r(inner radius)'],pressure_profil['Pr2'],label='12 hour',color='r')
plt.plot(pressure_profil['r(inner radius)'],pressure_profil['Pr3'],label='24 hour',color='b')
plt.grid()
plt.xlabel('RADIUS')
plt.ylabel('PRESSURE')
plt.xscale('log')
plt.title('pressure profile around the wellbore as function of time')
plt.legend()
```
